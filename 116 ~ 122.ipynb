# 117.조건부 확률과 곱셈법칙
* ① 조건부 확률의 정의
    * 사건A가 주어졌을 때, B의 조건부 확률은
    * P(A|B) = $ P(A∩B) \over  P(AB) $ = 단 P(B)>0 P(B)
* ② 독립법칙
    * 조건부 확률의 정의로부터 다음의 곱셈법칙이 성립 P(A) > 0,P(B) > 0이고 P(A∩B) = P(A|B)P(B)
* ③ 독립사건
    * 사건 A와 사건B가 독립이면 P(A∩B) = P(A)×P(B), P(A|B)=P(A), P(B|A) = P(B)
    * 결국 독립사건은 한 사건의 결과가 다른 사건에 영향을 주지 않을 때 A와 B를 독립사건이라고 한다.

---

해설 :  첫 번째 꺼낼 제품이 불량품일 사건을 A, 두 번째 꺼낸 제품이 불량품일 사건을 B라 하면 구하는 확률은 P(A∩B) = P(B|A)P(A) = $ 19 \over 99 $x $20 \over 100 $ = $19 \over 495$

# 118.확률변수
* 확률변수(random variable)
    * 표본공간 내에 있는 각 원소에 하나의 실수값을 대응 시키는 함수
    * ① 이산형 확률변수(Discrete random variable) :
        * 확률변수가 취할 수 있는 값이 유한하거나 또는 무한히 많더라도 하나씩 셀수 있 는 경우 (주로 계수자료 (count data))
    * ② 연속형 확률변수(Continuous random variable) :
        * 확률변수가 연속적인 구간 내의 값을 취하는 경우(높이, 무게, 온도, 거리, 주로 측정자료)
    * ③ X가 확률분포 f(x)를 갖는 확률변수라 하면, 기댓값은 아래와 같음
        *  X가 이산적 확률변수의 기댓값 E(X) = ∑xf(x)
        *  X가 연속적 확률변수의 기댓값 E(X) = ∫xf(x)

#### 용어정리
* 표본공간
    * 한 실험에서 나타날 수 있는 가능한 모든 결과 집합 S이다.
    * 즉, 시행에 의하여 일어날 수 있는 가능한 모든 결과로 된 집합을 말한다.

# 120.카이제곱 검정
* 카이제곱 검정은 범주별로 관측빈도와 기대빈도의 차이를 통해서 확률 모형이 데이터를 얼마나 잘 설명하는지 검정하는 통계적 방법이다.
* 카이제곱 검정 유형은 교차분할표 이용 여부에 따라 크게 일원 카이제곱 검정그 이원 카이제곱 검정으로 분류된다.
* ① 일원 카이제곱 검정
    * 적합성 검정(Goodness of fit test)
        * 모집단이 특정한 분포를 따른다는 가설에 대해 표본의 도수분포를 이용한 검정(가설검정 사례)
        * 귀무가설: 세 후보의 지지율이 같다.
        * 대립가설: 세 후보의 지지율이 다르다.</br></br>
     
* ② 이원 카이제곱 검정
    *  독립성 검정(Test of independence):
        * 한 모집단에서 표본을 추출하였을 때 두 이산형 변량에 대한 독립성 여부를 조사하는 것이다. (가설검정 사례)
        * 귀무가설: 두 변량이 서로 독립이다.
        * 대립가설: 두 변량이 서로 관련이 있다.
    * 동질성 검정 :
        * 두 집단 이상에서 각 범주 간의 비율이 서로 동일한지 검정하는 방법
        * 귀무가설: 모든 표본들의 비율은 동일하다.
        * 대립가설: 모든 표본들의 비율은 동일하지 않다.
* → 예를 들면 두 집단의 분포가 동일한가 등을 검정

# 121.중심극한정리
*  중심극한정리란 확률변수 X의 모집단분포가 정규분포가 아니더라도 표본크기가 충분히 크면, 랜덤표본의 평균의 표본분포는 근사적으로 정규분포 N(µ, $σ2 \over n $)을 따른다.
* 중심극한정리 이론으로 인해 모집단 분포가 균등분포, 이항분포, 지수분포를 따르더라도 표본크기가 상당히 크면 모집단의 특성을 추정하는 데 정규분포의 이점을 활용할 수 있다.

# 122.중요한 표본분포
* ① 카이제곱분포
    * x²분포는 t분포와 정규분포와는 달리 좌우대칭이 아니며 오른쪽으로 긴 꼬리를 갖는다.
    * 확률변수 x²은 제곱합의 합으로 구하기 때문에 음수는 가질 수 없고 다만 가장 왼쪽에서는 0의 값을 갖는다.
    * 한 모집단의 분산에 대한 추정과 검정을 함에 있어서 x² 분포를 이용한다.</br></br>
* ② t분포
    * t 분포는 표준정규분포보다 큰 분산을 갖는다. 그 이유는 모표준 편차 대신 표본표준편차 s를 사용하기 때문이다.
    *  모집단의 평균과 표준편차를 모르고 표본크기가 작은 경우에 모평균 µ에 대한 신뢰구간의 설정은 t통계량을 이용해야 한다.</br></br>
 
* ③ F분포
    * 오른쪽 꼬리분포이다. 자유도 (n₁-1)과 (n₂-1)이 증가할수록 F분포는 정규분포에 접근한다.
    * 두 모집단 분산의 동일성에 대한 추정과 검정을 하기 위하여 F분포 (F distribution)를 이용한다.
    * F분포는 두 모분산을 비교하는 데 사용할 뿐만 아니라 분산분석과 회귀분석을 위해서도 사용되는 중요한 확률분포이다.</br></br>
