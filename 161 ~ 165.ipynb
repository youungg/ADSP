# 161.인공신경망과 SOM의 차이
* ① SOM은 단순구조. 한 개의 입력층과 한 개의 출력층
* ② 출력 뉴런들은 승자 뉴런이 되기 위해 경쟁하고 오직 승자만이 학습하여 경쟁학습
* ③ SOM은 역전파 알고리즘이 아닙니다.
* ④ 비지도학습
#### 출제유형
* SOM과 비교해서 문제로 출제

# 162.인공신경망의 은닉층과 은닉 노드수를 정할 때 고려사항
* ① 다층신경망은 단층신경망에 비해 훈련이 어렵다.
* ② 노드가 많을수록 복잡성을 잡아내기 쉽지만, 과적합 가능성이 높다
* ③ 은닉층의 노드가 적으면 복잡한 의사결정 경계를 만들 수 없다.
* ④ 출력층 노드의 수가는 출력범주의 수로 입력의 수는 입력 차원의 수로 결정한다.
#### 출제유형
* 은닉층과 은닉수 고려사항 출제

# 163.인공신경망의 단점
* ① 결과에 대한 해석이 쉽지 않다
* ② 최적의 모형을 도출하는 것이 상대적으로 어렵다.
* ③ 데이터 정규화를 하지 않으면 지역해에 빠질 위험이 있다.
* ④ 모형이 복잡하면 훈련 과정에 시간이 많이 소요된다.
#### 출제유형
* 독립변수의 교호작용 쉽게 파악이 가능한 것은 의사결정나무

# 164. 인공신경망 구축 시 고려사항 3가지
* ① 입력변수
    * 신경망 모형은 복잡성으로 인하여 입력 자료의 선택에 매우 민감 즉 연속형 변수 또는 범주형 변수의 특성에 따라 좋지 않은 결과 도출
* ② 가중치 초기값
    * 역전파 알고리즘의 초기값에 따라 결과가 많이 달라진다.
    * 일반적으로 초기값은 '0'근처로 랜덤하게 선택함
    * 가중치가 0에 가까울수록 선형모형에 유사하게 됨
* ③ 과대적합(Overfitting)
    * 신경망에서는 많은 가중치를 추정해야 하므로 과대적합 문제가 자주 발생함
#### 출제유형
* 인공신경망 구축 시 고려사항이 출제

# 165. 인공신경망 활성화 함수
* 계단함수 (Step Function)
    *  임계값 0.1을 기준으로 활성화 또는 비활성화가 된다.</br></br>
* 부호함수 (Sign Function)
    * 임계값을 기준으로 양의 부호(+1) 또는 음의 부호(-1)를 출력</br></br>
* 시그모이드 함수 (Sigmoid Function)
    * 로지스틱 함수라고 하며 특정 임계값을 기준으로 출력값이 급격하게 변하는 계단함수와 달리 완만한 곡선 형태로 0~1사이의 값을 출력한다.</br></br>
* Tanh 함수 (Tanh Function)
    * 하이퍼볼릭 탄젠트함수
    * 확장된 시그모이드 함수
    * -1 ~ 1사이의 값을 출력한다.</br></br>
* ReLU (Rectified Linear Unit)
    * 입력값이 0보다 작으면 0을, 0보다 크면 입력값을 그대로 출력하는 함수</br></br>
* 소프트 맥스 함수 (Sigmoid Function)
    * 소프트맥스 함수는 목표값이 다범주인 경우에 사용
    * 입력받는 값을 정규화하여 0~1사이의 값으로 출력
    * 소프트맥스함수를 적용한 노드의 출력 값은 항상 1이다.</br></br>

#### 출제유형
* 각 활성화 함수의 특징이 출제
