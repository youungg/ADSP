# 130.다중공선성
* 회귀분석에서 사용된 모형의 일부 설명변수가 다른 설명 변수 간에 강한 상관관계가 나타나는 문제이다.
* 이는 회귀분석의 전제 가정을 위배하는 것으로 적절한 회귀분석을 위해 해결해야 하는 문제가 된다.
#### 출제유형
* 다중공선성 단답형 출제

# 131.다중공선성 진단방법
* ① vif 값이 10이 넘는다.
* ② 결정계수값이 크고 각 회귀계수들이 유의하지 않으면(p값이 크면) 다중공선성을 의심한다.
* ③ 독립변수간의 상관계수를 구한다

# 132.상관관계가 높은 변수들이 포함되었을 때 해결방안 (다중공선성)
* ① 중요하지 않는 변수 제거
* ② 해당변수의 상관관계에 따라 변수 통합
* ③ 자료 부족이 원인일 경우, 자료 보완
* ④ 능형회귀, 주성분 회귀 사용
#### 출제유형
* 다중공선성 해결방안 출제

# 133.정규화 선형회귀
* 정규화 선형회귀는 결국 특정변수의 과다한 기여를 제한하는 것.
* 그래서 상관된 다른 변수도 함께 모형에 투입이 가능
* 결국 → 회귀계수 베타에 제약 가함
* 능형회귀는 베타계수의 합이 특정 수치 이하로, 라쏘회귀는 베타계 수의 절대값이 합이 특정값 이하로 제약(패널티)을 준다.
* 이를 각각 L2 Norm(릿지), L1 Norm(라쏘)라 한다.
# 출제유형
* 객관식 문제 출제
    * 다중공선성의 해결방안으로 릿지(능형) 회귀를 이용한다.
    * 주성분분석도 다중공선성이 해결방안 중 하나이다.(O)

# 134.회귀분석 결과 개별회귀계수 t값의 의미
* ① t값은 독립변수와 종속변수간의 선형성의 강도를 의미한다.
* ② t값의 수식은 = 회귀계수/표준오차
#### 출제유형
* R의 다중회귀분석 결과를 통한 해석 문제 출제

# 135.회귀분석의 가정조건
* ① 선형성(독립변수의 변화에 따라 종속변수도 변화하는 선형)
* ② 독립성(잔차와 독립변수의 값이 관련되어 있지 않다.)
* ③ 등분산성(오차항들의 분포는 동일한 분산을 갖는다.)
* ④ 비상관성(잔차들끼리 상관이 없어야 한다.)
* ⑤ 정상성 또는 정규성(잔차항이 정규분포를 이뤄야 한다.)
#### 출제유형
* 회귀분석의 가정조건과 정의를 묻는 문제 출제
