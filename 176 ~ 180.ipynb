# 179.계층적 군집 연속형 거리개념
* 수학적거리
|종류|공식|두 군집 사이의 거리|
|:-:|:-:|-|
|유클리드</br>(Euclidean)|$d(i,j) = \sqrt { \underset {f=1} {{\overset{p}{\sum}}} (x_{if} - x_{jf})^2 }  $| 두 점간 차를 제곱하여 모두 더한 값의 양의 제곱근|
|맨해튼 </br>(Manhattan)|$d(i,j) =  { \underset {f=1} {{\overset{p}{\sum}}} |x_{if} - x_{jf} |}  $|• 시가 (City-block)거리라 고도 불림</br>• 두점간 차의 절댓값을 합한 값|
|민코프스키 </br>(Minkowskii)|$d(i,j) = [ { \underset {f=1} {{\overset{p}{\sum}}} |x_{if} - x_{jf}| ^ m}]^{1 \over m}  $|• m차원 민코프스키 공간에서의 거리</br>• m = 1일 때 맨해튼 거리와 같음</br>• m = 2일 때 유클리드 거리와 같음|
|표준화 </br>(Standardized)|$d(i,j) = \sqrt {  (x_{i} - x_{j}) D^{-1} (x_{i} - x_{j}) } $</br>$D = D_{iag}(S_{11},···,S_{pp}) $ : 표본분산(대각) 행렬|• 변수의 측정단위를 표준화한 거리|
|마할라노비스</br>(Mahalanobis Distance)|$d(i,j) = \sqrt {  (x_{i} - x_{j}) S^{-1} (x_{i} - x_{j}) } $</br>$S = (S_{ij})_px_p $ : 표본공분산(대각) 행렬|• 변수의 표준화와 함께 변수간의 상관성(분포형태)을 동시에 고려한 통계적 거리|

# 180. 기타 거리개념
* ① 명목형 거리
|거리|정의|
|:-:|-|
|단순 일치 계수</br>(Simple Matching Coefficient)|• $매칭된 속성의 개수 \over 속성의 개수 $ </br>• 전체 속성 중에서 일치하는 속성의 비율|
|자카드계수</br>(Jaccard)|•$J(A,B) = {|A \cap B| \over |A \cup B|} = {|A \cap B| \over |A|+|B|-|A \cap B|}$ </br>•두 집합 사이의 유사도를 측정하는 방법 </br>• 0과 1 사이의 값을 가지며 두 집합이 동일하면 1의 값, 공통의 원소가 하나도 없으면 0의 값을 가짐|
* ② 순서형 거리
|거리|정의|
|:-:|-|
|순위상관 계수|• 순위상관계수(Rank Correlation Coefficient)를 이용하여 거리를 측정한다. </br>• $\rho = 1 - {G \underset{i=1} {\overset{n}{\sum} d_i^2} \over n(n^2-1) }$  </br>• $ \chi_i$ :관측치  </br>•$d_i$ : i번째 데이터 순위 차|
* ③ 캔버라 거리
    * 가중치 있는 맨해튼 거리.
    * 원점 주변에 흩어져 있는 데이터에 주로 사용
    * $ d(x,y) = {\underset{i=1}{\overset{n}{\sum}} |x_i - y_i| \over (x_i + y_i)} $
 
* ④ 체비셰프 거리
    * 두 관찰치 사이의 거리 측정으로 변숫값의 최대 차이의 절댓값 d(x,y) = maxlx, -y
* ⑤ 코사인 유사도
    * 두 벡터의 내적을 각 벡터의 크기로 나눈 값을 1에서뺀 것
    * 코사인 유사도는 두 벡터 사이의 각도를-1부터 1사이의 값으로 나타내는 척도
    * 두 벡터가 같은 방향(두 벡터 사이의 각도 0)이면 1, 두 벡터가 완전 반대 방향이면-1이 된다.
    * 값이 1인 경우 벡터 간의 유사도가 크다.
    * -1인 경우 유사도가 매우 작음을 의미한다.
    * similarity= $ A · B \over \Vert A \Vert\Vert B \Vert $
