# 172.군집분석 타당도 지표
* 비지도학습의 특성상 어떠한 지표라도 정확하게 성능을 평가하기는 어렵지만, 군 집화의 성능을 평가하는 대표적인 방법으로는 실루엣 계수와 Dunn Index가 있다.
    * ① 실루엣(Silhouette)이 1이면 한 군집의 모든 개체가 한치도 떨어져 있지 않고 붙어 있는 경우. 보통 실루엣 지표가 0.5보다 크면 군집 결과가 타당한 것으로 평가. 숫자 1에 가까울수록 군집화가 잘되었다고 판단함.
    * ② Dumm Index는 (군집과 군집 사이 거리 중 최솟값)/(군집 내 데이 터들 거리 중 최댓 값)으로 분자가 클수록 군집 간 거리가 멀고, 분모 값이 작을수록 군집 내 데이터가 모여 있다. Dumm Index가 클수록 군집화가 잘 되었다고 볼 수 있음.

# 173.계층적 군집의 거리


|군집방법|두 군집 사이의 거리|
|:-:|-|
|단일연결법 </br>(single linkage)|• 한 군집의 점과 다른 군집의 점 사이의 가장 짧은 거리(shortest distance)</br>• 사슴 모양으로 생길 수 있으며, 고립된 군집 을 찾는데 중점을 둔 방법이다.|
|완전연결법 </br>(complete linkage)|• 두 군집 사이의 거리를 각 군집에서 하나 관측값을 뽑았을 때 나타날 수 있는 거리 최댓값을 측정한다. </br>• 같은 군집에 속하는 관측치는 알려진 거리보다 짧으며, 군집들의 내부 응집성에 중점을 둔 방법이다.|
|평균연결법 </br>(average linkage)|• 모든 항목에 대한 거리 평균을 구하면서 군집 화를 하기 때문에 계산량이 불필요하게 많아 질 수 있다.|
|중심연결법 </br>(centroid)|• 두 군집이 결합할 때 새로운 군집의 평균은 가중평균을 통해 구해진다.|
|와드연결법 </br>(ward linkage)|• 군집 내의 오차제곱합에 기초하여 군집|

# 174. 비계층적 군집분석
* ① k-평균 군집(k-means clustering)
    * k-평균 군집은 원하는 군집 수만큼 초기값을 지정하고, 각 개체를 가까운 초기값에 할당 하여 군집을 형성한 뒤, 각 군집의 평균을 재계산하여 초기값을 갱신한다.
    * 갱신된 값에 대해 위의 할당 과정을 반복하여 개의 최종군집을 형성하는 방법이다.
* ② k-평균 군집 Process
    * 단계 1: 초기 군집 중심(Centroid)으로 k개의 객체를 임의로 선택 한다.
    * 단계 2. 각 자료를 가장 가까운 군집 중심에 할당한다. 즉 자료들의 군집의 중심점(평균) 으로부터 오차 제곱합이 최소가 되도록 각 자료를 할당한다.
    * 단계 3: 각 군집 내의 자료들의 평균을 계산하여 군집의 중심을 갱신한다.
    * 단계 4: 군집 중심의 변화가 없을 때까지 단계 2와 단계 3을 반복한다.
* ③ k-means 군집 단점
    * 잡음이나 이상값에 영향을 받기 쉽다. 평균 대신 중앙값을 사용 하는 k-medoids(중앙값) 군집을 사용할 수 있다. 또한 k-mean 분석 전에 이상값을 제거하는 것도 좋은 방법이다.
    * 계층적 군집과는 달리 k-mean 군집은 사전에 군집의 수를 정해 주어야 한다.

# 175.혼합분포군집
