# 105. 벡터의 연산 (벡터들 길이가 동일하지 않은 경우)
* 두 벡터가 원소가 개수가 다르더라도 연산과정에서 원소의 개수가 적은 쪽의 벡터는 원소 개수가 많은 쪽의 벡터와 동일하게 원소의 개수를 맞춘다.
    * ex)  예 x<-c(1,2,3) y<-c(1,2,3,4,5,6)
    * x+y의 값은 2 4 6 5 7 9
