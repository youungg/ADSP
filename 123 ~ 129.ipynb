# 123.가설검정
* 모집단분포를 포함해서 모집단에 대한 통계적 추론을 실시하기 위해서는 추정 과 가설검정을 이용한다.
* 추정에 있어서 추론이 표본으로부터 모집단으로 진행하여 모수의 참값을 추정하는 과정을 거치는 반면 가설검정에 있어서는 미지의 모수에 대한 주장 또는 가정의 타당성 여부를 표본결과로 검정하는 과정을 거친다.
* ① 검정의 오류
    * 가설검정과 관련된 오류에는 두 가지가 있다. 귀무가설 H가 실제로는 에도 불구하고 허위라고 결론을 내릴 오류를 제I종 오류(type I error : α로 표시한다.
    *  이러한 오류를 범할 확률을 검정의 유의수준(level of significance) 위험 수준(level of risk)이라 한다.</br></br></br>
 
||귀무가설 참|귀무가설 거짓|
|-|-|-|
|$H_0$채택|옳은 결정 확률=1-α|제II 종 오류 확률 = β|
|$H_0$기각|제1종 오류 확률 = α|옳은 결정 확률 = 1- β|

* 귀무가설을 채택하느냐 또는 기각하느냐에 따라 반드시 (α+β)만큼의 오차가 발생하게 되므로 검정을 할 때 가급적 제I종오류와 제II종 오류는 작을수록 좋지만 표본크기가 일정한 경우에는 동시에 감소시킬 수 없다.
* 표본크기가 일정할 때 제1종 오류를 범할 확률을 감소시키면 제II종 오류를 범할 확률은 증가한다. 이와 같이 α 와 ẞ는 부의 관계이다.
* α위험이 위험보다 더욱 심각하기 때문에 오류를 통제하는 것이 일반적이다. α의 값은 보통 1%, 2%, 5%, 10% 등으로 의사결정자 가 결정</br></br>

* ② 가설검정의 순서
    *  가설의 설정
    *  유의수준 의 결정
        * 검정통계량을 사용할 때
            * 유의수준 에 해당하는 임계치 및 기각영역의 결정
            * 검정통계량의 계산
            * 의사결정
        * p값을 사용할 때
            * 검정통계량의 계산
            * p값의 계산</br></br>
         
* ③ 검정력
    *  제II종 오류를 범할 확률 ẞ는 대립가설이 사실이기 때문에 당연히 귀무를 기각해야 함에도 불구하고 이를 채택하는 확률을 말한다.
    * 따라서 (1-ẞ)로 나타내는 검정력(Power of test)은 거짓인 귀무가 기각하는 확률을 말한다.
         
#### 용어정리
* 검정통계량
    * 검정통계량이란 귀무가설의 진위여부를 규명하기 위하여 가설검정과 관련된 모집단으로부터 표본을 무작위로 추출하여 필요한 계산을 하는 표본통계량을 말한다.

#### 출제유형
* 1종오류와 2종오류 개념문제 출제
* "가설검정에는 제2종 오류의 크기를 0.1, 0.05 등으로 고정시킨 뒤 기각역을 설정한다." 오답보기로 "제1종 오류 고정한다." 올바른 표기이다.
* "표본크기가 일정하다면 제1종오류와 제2종오류는 동시에 감소시킬 수 있다."는 잘못된 오답보기로 올바른 표현은 "제 1종오류와 제2종 오류는 동시에 감소가 불가능하다."


# 124.유의확률 vs 유의수준
* 유의수준: 제 1종 오류를 범할 확률의 최대값을 ∝라 표기하고 우리는 이 α값을 유의수준(Significance level)이라 한다.
* 유의확률(p값)은 귀무가설(Null hypothesis)이 맞다는 전제하에 표본에서 실제로 관측된 통계치와 '같거나 더 극단적인' 통계치가 관측될 확률이다.
#### 출제유형
* 만약 p값이 0.01 나왔다면 해석은?
    * → 귀무가설이 사실일 때 이러한 결과가 나올 확률이 1/100이다.
* 유의수준 0.05란 어떤 의미일까요?
    * → 동일한 검정법을 여러 번 반복하여 사용할 때 잘못으로 귀무가설을 기각하게 될 경우가 전체의 5% 이하일 것이라는 의미
* 유의확률이 작을수록 귀무가설에 대한 반증이 강하다.(o)
* 유의확률은 귀무가설 하에서 계산된 값이다.(o)

# 127.확률적 표본추출방법
* ① 단순 무작위추출(Simple random sampling):
    * 모집단의 각 개체가 표본으로 선택될 확률이 동일하게 추출되는 경우이다.
    * 모집단의 개체 수가 N 중에서 임의로 n(표본 수)개를 선택한다면, 개별 개체 가 선택될 확률은 모두 n/N으로 일정하다.
* ② 계통추출(Systematic sampling):
    * 모집단의 개체에 1, 2,..., N이라는 일련번호를 부여한 후, 첫 번째 표본 임의로 선택하고 일정 간격으로 다음 표본을 선택한다.
* ③ 층화추출(Stratified sampling):
    * 모집단을 성격에 따라 몇 개의 집단 또는 층(strata)으로 나누고, 각 집단 내에 서 원하는 크기의 표본을 무작위로 추출한다.
* ④ 군집추출(Cluster sampling):
    * 모집단을 특성에 따라 여러 개의 집단(duster)으로 나눈다.
    * 이들 집단 중에서 몇 개를 선택한 후, 선택된 집단 내에서 필요한 만큼의 표본을 임 선택한다.

# 128. 비모수 검정

* 최소한의 가정을 전제로 하므로 가정이 만족하지 않음으로 발생하는 오류를 줄일 수 있다.
* 모수 검정은 등간척도와 비율척도로 측정된 자료를 분석한다.
* 비모수 검정은 명목척도와 순서척도로 측정된 자료를 분석한다.
