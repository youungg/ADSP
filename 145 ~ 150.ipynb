# 145.데이터 마이닝 6가지 분석기법(기능)
* ① 분류(Classification)
    * 분류는 새롭게 나타난 현상을 검토하여 기존의 분류, 정의된 집합에 배정하는 것을 의미한다. 분류기법으로는 의사결정나무(Decision tree), memory-based reasoning 등이 있다.</BR></BR>
* ② 추정(Estimation)
    * 추정은 연속된 변수의 값을 추정하는 데 사용한다. 즉 주어진 입력데이터를 사용하여 알려지지 않는 결과의 값을 추정해낸다.
    * 추정 기법으로는 신경망 모형이 있다.</BR></BR>
* ③ 예측(Prediction)
    * 예측은 미래의 양상을 예측하거나 미래의 값을 추정한다는 것을 제외하면 분류나 추정과 동일한 의미를 갖는다. 입력 데이터의 성격에 따라 장바구니 분석, 의사결정나무, 신경망 등이 예측 기법으로 사용된다.</BR></BR>
 
* ④ 연관분석(Association Analysis)
    * 연관분석은 '같이 팔리는 물건'같이 아이템의 연관성을 파악하는 분석이다. 대표적 분석기법으로 장바구니 분석이 있다.</BR></BR>
* ⑤ 군집(Clustering)
    * 군집과 분류와의 차이점을 보면, 군집은 미리 정의된 기준이나 예시에 의해서가 아닌 레코드 자체가 지니고 있는 다른 레코드와의 유사성에 의해 그룹화되고 이질성에 의해 세분화된다.
    * 군집은 데이터마이닝이나 모델링의 준비 단계로서 사용된다.</BR></BR>
 
* ⑥ 기술(Description)
    * 데이터가 가지고 있는 의미를 단순하게 기술하는 것도 의미를 파악하는 것만큼 중요하다. 훌륭한 기술은 데이터가 암시하는 바에 대해 설명이 가능해야 하며, 또한 설명에 대한 답을 찾아낼 수 있어야한다.</BR></BR>
#### 출제유형
* 분석기법의 개념 객관식 및 단답형 출제

# 147.분류분석 vs 예측분석 차이점
* ① 공통점: 레코드 특성 속성값 미리 맞힘.
* ② 차이점
    * 분류: 레코드 범주형 속성의 값
    * 예측: 레코드 연속형 속성의 값
#### 출제유형
* 신용평가 대상자를 위험에 따라 상·중·하로 분류하고 새로운 고객이 왔을 때 어떤 위험군에 속할지 예측하는데 활용할 수 있다(X). (예측 아니고 분류에 해당된다.)

# 148.지도학습 VS 비지도학습
* ① 목표변수의 존재 유무에 따라 구분한다.
* ② 사례로 판단하기
    * 과거 소비자들의 인구통계학적, 금융데이터 바탕으로 새로운 소비자들의 대출 여부를 결정한다.(대출여부 0,1)
    * 온라인 서점에서 기존 구매 고객들의 구매 패턴을 파악하여 책을 추천하다. (고객들의 구매 패턴의 그룹핑)
    * 기존 파산회사와 파산하지 않은 회사들의 재정상태 데이터를 기반으로 회사의 파산 여부를 예측한다.(파산여부 0,1)→ 지도학습
* ③ 비슷한 성향을 가진 고객군을 파악하여 고객 타켓팅에 활용
    * → 목표변수가 없어 비지도학습
#### 출제유형
* 사례를 통한 지도학습과 비지도학습 구분 출제. 분석기법에서 비지도학습은 군집분석, 연관성분석, 주성분분석, SOM이 있다.

# 149.모형평가
* ① 분류 모형의 평가에 사용되는 그래프로 x축은 (1-특이도) y축은 민감도로 그려지는 것은 무엇인가? ROC 그래프
* ② ROC 그래프의 밑부분 면적이 넓을수록 좋은 모형이다.(x=0,y=1)
    * AUC = 0.5 분류 능력이 없음을 의미함
* ③ 데이터 분할
    *  데이터를 일반적으로 구축용(train), 검정용(validation), 시험용 (test)으로 분리한다.
    * 검정용 데이터는 구축된 모델의 과잉 또는 과소맞춤 등에 대한미세조정 절차를 위해 사용한다.
    * 시험용 데이터는 성능평가를 위해 사용한다.
#### 출제유형

# 150.교차검증(Cross Validation)
* 교차검증은 모델의 일반화 오차에 대해 신뢰할만한 추정치를 얻기 위해 훈련, 평가 데이터를 기반으로 모형을 평가하는 방법</BR></BR>
* ① 홀드아웃(HoldOut)
    * 주어진 원천 데이터를 랜덤하게 두 분류로 분리하여 교차검정을 실시하는 방법으로 하나는 모형의 학습 및 구축을 위한 훈련용 자료로 하나는 성과 평가를 위한 검증용 자료로 사용한다.
* ② K-Fold 교차검증
    * 데이터 집합을 무작위로 동일 크기를 갖는 k개의 부분 집합으로 나누고 그 중 1개 집합을 평가용 데이터로 나머지를 k-1개 집합을 학습데이터로 선정하여 모형 평가하는 방법.
* ③ 붓스트랩
    * 평가를 반복한다는 측면에서 교차검증과 유사하나 훈련용 자료를 반복 재선정한다는 점에서 차이가 있다.
    * 즉 붓스트랩은 관측치를 한 번 이상 훈련용 자료로 사용하는 복원추출법에 기반한다.
    * 이로 인해 특정 샘플이 학습데이터에 포함될 확률은 63.2% 선택되지 않을 확률은 36.8%
#### 출제유형
* 교차검증의 종류이 개념이 단답형 및 객관식 출제
